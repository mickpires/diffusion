```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = .835
def u(x,t):
    global k
    somatorio = 0
    for n in range(1,1000):
        # print(somatorio)
        somatorio += 100*((-1)**n - 2)/(n * np.pi) * np.sin(n*np.pi*x/10)*np.exp(-k*(n* np.pi/10)**2*t)
    return 100 - 5*x + somatorio
```

```python
t = np.linspace(0,100,1000)
plt.plot(t,u(1.1,t))
plt.xlabel('tempo')
plt.ylabel('densidade de difusão no ponto 1.1')
plt.title('Solução analitica u(1.1,t)')
```
